# Autodifferentiation

```python
torch.manual_seed(0)
X = torch.randn((2,2),requires_grad=True)
A = torch.rand_like(X)
f_X = 0.5*torch.linalg.norm(A-X)**2

print("f value:", f_X)
print("gradient X:",X.grad)
```

```python
make_dot(f_X,params={"X":X})
```

```python
f_X.backward()
print("gradient X:\n",X.grad)
```

```python
X = torch.randn((2,2),requires_grad=True)
Y = torch.randn((2,2),requires_grad=True)
A = torch.rand_like(X)
B = torch.rand_like(Y)

f_XY = 0.5*torch.linalg.norm(A@X + B@Y)**2
```

```python
make_dot(f_XY,params={"X":X,"Y":Y})
```

```python
grad_X = torch.autograd.grad(f_XY,X)[0]
X.grad = grad_X
```

```python
step_size = 0.1
n_iter = 100
A = torch.rand((2,2))
def loss(X):
    return 0.5*torch.linalg.norm(A-X)**2
X = torch.randn((2,2),requires_grad=True)
for i in range(n_iter):
    l = loss(X)
    l.backward()
    with torch.no_grad():
        X -= step_size * X.grad
        X.grad.zero_()
    if i % 10 == 0:
        print(f"iter {i}: loss={l.item():.4f}")
```

```python
import torch.nn as nn
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out
```
