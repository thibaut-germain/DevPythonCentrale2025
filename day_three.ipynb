## Visulization (matplotlib)


````python
import numpy as np
import matplotlib.pyplot as plt


# process gaussien
np.random.seed(0)
n_samples = 100
sigma = 0.1
time = np.linspace(0,1,n_samples)
traj_1 = time + sigma * np.random.randn(n_samples)
traj_2 = 1-time + sigma * np.random.randn(n_samples)
````

````python
plt.plot(traj_1)
````

````python
fig = plt.figure(figsize=(10,2))
plt.plot(time,traj_1,label="traj_1")
plt.xlabel("time")
plt.ylabel("value")
plt.title("My title")
````

````python
fig = plt.figure(figsize=(10,2))
plt.plot(time,traj_1,label="traj_1")
plt.plot(time,traj_2,label="traj_2")
plt.xlabel("time")
plt.ylabel("value")
plt.legend()
plt.title("My title")
plt.tight_layout()
plt.savefig("figure/my_first_figure.pdf")#,bbox_inches='tight') #.png, .svg, .pdf
````

````python
fig = plt.figure(figsize=(10,2))
plt.plot(time,traj_1,label="traj_1")
plt.plot(time,traj_2,label="traj_2")
plt.xlabel("time")
plt.ylabel("value")
plt.legend()
plt.title("My title")
plt.tight_layout()
plt.savefig("figure/my_first_figure_2.png", dpi = 200, bbox_inches='tight') #.png, .svg, .pdf
````

````python
fig,axs = plt.subplots(2,1,figsize=(10,4))
axs[0].plot(time,traj_1)
axs[0].set_xlabel("time")
axs[0].set_title("traj_1")
axs[1].plot(time,traj_2)
axs[1].set_xlabel("time")
axs[1].set_title("traj_2")
````

````python
fig,ax = plt.subplots(figsize=(10,2))
ax.plot(time,traj_1,label="traj_1")
ax.plot(time,traj_2,label="traj_2")
ax.set_xlabel("time")
ax.set_title(r"Gaussion process with $X_t \sim \mathcal{N}(0,0.1)$")

````

````python
np.random.seed(1)
n_samples = 30
x = np.random.randn(n_samples)
y = np.random.randn(n_samples)
importance_weight = (np.random.rand(n_samples)+1)*100
classes = np.random.randint(0,4,n_samples)


fig = plt.figure(figsize=(4,4))
plt.scatter(x,y,s=importance_weight,c=classes)
plt.tight_layout()
````

````python
import  plotly.graph_objects as go

classe_name = np.unique(classes)


fig = go.Figure()
for idx in classe_name: 
    mask = classes == idx
    xf = x[mask]
    yf = y[mask]
    wf = importance_weight[mask]/10
    fig.add_trace(
        go.Scatter(x=xf,y=yf,mode="markers",marker=dict(size=wf))
    )

fig
````

````python
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=x,y=y,mode="markers",marker=dict(color = classes,size = importance_weight/10))
)
````
