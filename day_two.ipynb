```python
def label_fct(id:int,filename:str,format:str="csv")->str: 
    """ Defined filename
    Args:
        id (int): experiment
        filename (str): experiment name
        format (str, optional): result format. Defaults to "csv".

    Returns:
        str: filename
    """
    return f"{filename}_{id}."+format
```
